```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Feature Engineering - create a column 'age_group'
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age', 'age_group']].head()
```

```python
# create column 'purchase_frequency_days'

frequency_map = {'Fortnightly': 14,
                 'Weekly': 7,
                 'Monthly': 30,
                 'Quarterly': 90,
                 'Bi-weekly': 14,
                 'Annually': 365,
                 'Every 3 Months': 90}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_map)
```

```python
from sqlalchemy import create_engine
import urllib
```

```python
# --- Update these with your actual values ---
server = r"DESKTOP-G1B3DCM\SQLEXPRESS"        # e.g. "LAPTOP123\SQLEXPRESS" or "localhost"
database = "customer_behavior"     # e.g. "TestDB"
table_name = "customer_behavior_table"      # name of table to create or append to
```

```python
# Create connection string for Windows Authentication
params = urllib.parse.quote_plus(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
)
```

```python
# Build the engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
# Push the DataFrame to SQL Server
df.to_sql(table_name, con=engine, if_exists='replace', index=False)

print("Data pushed to SQL Server successfully!")
```
